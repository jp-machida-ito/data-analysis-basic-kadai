```python
#課題005

import matplotlib.pyplot as plt
import numpy as np

#6領域のサブプロットを作成
fig,axes = plt.subplots(2,3)

#折れ線グラフ
y0 = [0,1,-5,8]
axes[0,0].plot(y0)

#sin関数
x1 = np.linspace(0,10,100) #0から10までの幅で配列を100個作成
y1 = 2 + 2*np.sin(2*x1)
axes[0,1].plot(x1,y1)

#ヒストグラム
x2 = np.random .normal(15,5,2000) #平均15、標準偏差5の標準正規分布から2000個の乱数の生成
axes[0,2].hist(x2,bins=30)

#散布図
np.random.seed(42)
x3 = 4+np.random  .normal(0,2,60)
y3 = 4+np.random  .normal(0,2,len(x3))
axes[1,0].scatter(x3,y3)

plt.tight_layout()
plt.show()
```
